```python
from cs103 import *
from typing import List,NamedTuple,Optional
from enum import Enum
from matplotlib import pyplot
import csv

###########
# Functions

@typecheck
def main(filename: str) -> None:
    """
    Reads the file from given filename, analyzes the data, returns the 
    result in the form of a pie chart like the one in step 1c.
    """
    # Template from HtDAP, based on function composition
    return analyze(read(filename))  

@typecheck
def read(filename: str) -> List[CountryData]:
    """    
    reads information from the specified file and returns a list of countries' data
    """
    #return []  #stub
    # Template from HtDAP
    # locd contains the result so far
    locd = [] # type: List[CountryData]

    with open(filename) as csvfile:
       
        reader = csv.reader(csvfile, delimiter=',')
        next(reader) # skip header line

        for row in reader:
            # you may not need to store all the rows, and you may need
            # to convert some of the strings to other types
            cd = CountryData(row[1], parse_float(row[2]), parse_float(row[3]), parse_govt_type(row[4]), parse_float(remove_commas((row[5]))))
            locd.append(cd)
    return locd

@typecheck
def remove_commas(s:str) -> str:
    """
    takes in a string and removes the commas in it.
    """
    #return "x" #stub
    #template based on atomic non-distinct
    
    return s.replace(",", "")

@typecheck
def parse_govt_type (s: str) -> Gov_Type:
    """
    Given a string which must be one of "Republic", "Constitutional monarchy", "Absolute monarchy", "n/a", or
    "Communist state", returns the corresponding Gov_Type (enumeration)
    """
    #return Gov_Type.Republic #stub
    #return ...(s) #template
    
    if s == "Republic":
        return Gov_Type.Republic
    elif s == "Constitutional monarchy":
        return Gov_Type.Constitutional_monarchy
    elif s == "Absolute monarchy":
        return Gov_Type.Absolute_monarchy
    elif s == "n/a":
        return Gov_Type.na
    elif s == "Communist state":
        return Gov_Type.Communist_state

@typecheck
def analyze(locd: List[CountryData]) -> None:
    """
    returns the average GDP for each government type and produces a pie chart for this
    """
    #return None #stub
    # template based on Visualization

    Gov_Type = gov_types
    average_gdp_per_gov_type = get_avg_gdp_per_gov_type(locd,Gov_Type)
    
    new_gts = []
    new_avgs = []
    for avg, gt in zip(average_gdp_per_gov_type, Gov_Type):
        if avg != 0:
            new_avgs.append(avg)
            new_gts.append(gt)
            
    pyplot.pie(new_avgs, labels= new_gts, autopct='%.0f%%', shadow = False)
    pyplot.show()
    return (None)  


@typecheck
def get_avg_gdp_per_gov_type(locd: List[CountryData], logt:List[Gov_Type]) -> List[float]:
    """
    returns a list of average gdp for each type of government
    """
    #return [] #stub
    #template from arbitrary-sized data w/ add'l parameter logt
    acc = [] #type:List[float]
    for gt in logt:
        acc.append(get_avg_gdp_for_gov_type(locd, gt))
    return acc

@typecheck
def get_avg_gdp_for_gov_type(locd: List[CountryData], gt: Gov_Type) -> float:
    """
    returns the average gdp for all countries in a list of countries for their given government types
    """
    #return 0.0 #stub
    #template based on function composition
    #Plan:
    # 1.) Filter the locd in order to obtain the country of a given government type
    # 2.) Find the average gdp of all countries in 1.)
    
    return get_average_gdp(calc_gdp(count(locd, gt)))

@typecheck
def get_average_gdp(lof: List[float]) -> float:
    """
    return the average gdp for all countries in the locd
    """
    # return 0.0 #stub
    # template taken from List[float]
   
    if len(lof) == 0:
        return 0
   
    # acc holds the sum of gdp in locd so far
    acc = 0 #type: float
   
    for f in lof:
        acc += f
    
    return acc/len(lof)

@typecheck
def calc_gdp(locd: List[CountryData]) -> List[float]:
    """
    returns the gdp for a list of countries by doing population * gdp per capita
    """
    #return [] #stub
    #template from List[CountryData]

    #gdp_acc contains result so far
    gdp_acc = [] #type: List[float]
   
    for cd in locd :
        gdp_acc.append (cd.population * cd.gdp_per_cap)

    return gdp_acc

@typecheck
def count(locd: List[CountryData], gt: Gov_Type) -> List[CountryData]:
    """
    takes in a list of country data and a government type, returns a count for this.
    that is, it takes in data from five countries and keeps count of how many of 
    these countries are of that government type
    """
    #return [] #stub
    #Template from List[CountryData] w/ add'l parameter
   
    #count has results seen so far
    count = [] #type: List[CountryData]
    for cd in locd:
        if is_govt_type(cd, gt):
            count.append(cd)
    return count

@typecheck
def is_govt_type(cd: CountryData, gt: Gov_Type) -> bool:
    """
    returns True if the government is either Republic, Constitutional monarchy, Absolute monarchy, n/a, or
    Communist state, False otherwise.The result of this function is being used in the function above
    """
    #return True #stub
    #template based on CountryData w/ add'l parameter
    return cd.gov_type == gt

#Begin testing
start_testing()

### PLEASE NOTE: we are not testing functions with an empty government type, as every country has
# a government and even if it didn't, it would technically fall in the n/a category

#examples for is_govt_type

expect(is_govt_type(CD1, Gov_Type.Communist_state), True)
expect(is_govt_type(CD2, Gov_Type.Absolute_monarchy), False)
expect(is_govt_type(CD5, Gov_Type.na), True)
expect(is_govt_type(CD8, Gov_Type.Constitutional_monarchy), False)

#examples for count

expect(count([], Gov_Type.Communist_state), [])
expect(count(L0, Gov_Type.Communist_state), [])
expect(count(L1, Gov_Type.Republic), [CD2, CD6, CD8, CD10])
expect(count(L1, Gov_Type.na), [CD5])
expect(count(L2, Gov_Type.Communist_state), [])
expect(count(L2, Gov_Type.Absolute_monarchy), [CD4])

# examples for calc_gdp 

expect(calc_gdp([]),[])
expect(calc_gdp(L1),[19143199000000, 
                     32626523930, 
                     1100715127400,
                     9391991844, 
                     90373290000, 
                     73784186622, 
                     1618303113000, 
                     2177711284247, 
                     29169522300, 
                     8011536400000])

# examples for get_average_gdp

expect(get_average_gdp([]),0)
expect(get_average_gdp([23.33,65.51,32.87]), 40.57)
expect(get_average_gdp([100.34,334.76,2214,6765.43]), 2353.6325)


# Examples for get_avg_gdp_for_gov_type

expect(get_avg_gdp_for_gov_type([],Gov_Type.na), 0)
expect(get_avg_gdp_for_gov_type(L1, Gov_Type.Republic), 2573914598699.75)
expect(get_avg_gdp_for_gov_type(L1, Gov_Type.Constitutional_monarchy),1359509120200.0)
expect(get_avg_gdp_for_gov_type(L2, Gov_Type.Republic),761373998266.3334)


# Examples for get_avg_gdp_per_gov_type

expect(get_avg_gdp_per_gov_type([],[]), [])
expect(get_avg_gdp_per_gov_type([],LG0), [])
expect(get_avg_gdp_per_gov_type(L0,[]), [])
expect(get_avg_gdp_per_gov_type(L0, LG0),[])
expect(get_avg_gdp_per_gov_type(L2, LG0),[])
expect(get_avg_gdp_per_gov_type(L1, LG1), [2573914598699.75, 
                                           19280757072.0, 
                                           1359509120200.0, 
                                           90373290000.0, 
                                           19143199000000.0])
expect(get_avg_gdp_per_gov_type(L2, LG1),[761373998266.3334, 9391991844.0, 0, 0, 0])

# Examples for remove_commas

expect(remove_commas(""), "")
expect(remove_commas(","), "")
expect(remove_commas("1000"), "1000")
expect(remove_commas("3,000"), "3000")

# Examples and tests for read

expect(read('test1.csv'),
        [CountryData('Albania', 19.3, 11305.0, Gov_Type.Republic, 2886026),
         CountryData('Algeria', 11.2, 14687.0, Gov_Type.Republic, 40400000),
         CountryData('Antigua and Barbuda', 11.0, 22968.0, Gov_Type.Constitutional_monarchy, 86295),
         CountryData('Armenia', 18.5, 8070.0, Gov_Type.Republic, 2995100),
         CountryData('Australia', 5.8, 45514.0, Gov_Type.Constitutional_monarchy, 24184100)])

expect(read('test2.csv'),
        [CountryData('Yemen', 35.0, 3785.0,  Gov_Type.Republic, 27478000),
         CountryData('Zambia', 15.0, 3904.0,  Gov_Type.Republic, 15933883),
         CountryData('Zimbabwe', 80.0, 1792.0,  Gov_Type.Republic, 14240168),
         CountryData('United States', 4.9, 53042.0,  Gov_Type.Republic, 319000000),
         CountryData('China', 4.1, 14107.0, Gov_Type.Communist_state, 1357000000)])

expect(read('test3.csv'),
        [])

# Examples for parse_govt_type

expect(parse_govt_type("Republic"), Gov_Type.Republic)
expect(parse_govt_type("Constitutional monarchy"), Gov_Type.Constitutional_monarchy)
expect(parse_govt_type("Absolute monarchy"), Gov_Type.Absolute_monarchy)
expect(parse_govt_type("n/a"), Gov_Type.na)
expect(parse_govt_type("Communist state"), Gov_Type.Communist_state)

# Examples for analyze function

expect(analyze(L0), None)
expect(analyze(L1), None)
expect(analyze(L2), None)

#Examples for main function

expect(main("country_per_cap_gdp_unemployment_gov_type_pop.csv"), None)
expect(main("test1.csv"), None)
expect(main("test2.csv"), None)
expect(main("test3.csv"), None)

# main("country_per_cap_gdp_unemployment_gov_type_pop.csv") 

# this has been commented out because of Professor Steve's 
# post on Piazza where he has stated that we should comment 
# out our main function calling the original .csv file

# show testing summary
summary()
```
